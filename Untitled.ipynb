```python
import os

files = os.listdir("./data/images")
files = [file for file in files if file[0] != "."]
```

```python
files
```

```python
import cv2

for file in files:
    imgs = os.listdir("./data/images/" + file)
    for img in imgs:
        
        cv2.imread("./data/images/" + file + "/" + img)
```
